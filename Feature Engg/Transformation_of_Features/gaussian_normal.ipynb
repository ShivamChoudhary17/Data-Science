```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/titanic.csv', usecols=['Age', 'Fare', 'Survived'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df.head()
```

```python
df.Age.fillna(df.Age.median(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import scipy.stats as stat
import pylab
```

```python

#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
# If blue line is on red then it is normally distributed
plot_data(df, 'Age')
```

```python
plot_data(df, 'Fare')
```

```python
plot_data(df, 'Survived')
```

## Logarithmic Transformation

```python
import numpy as np
df['Age_log'] = np.log(df['Age'])
plot_data(df, 'Age_log')
```

## Reciprocal Trnasformation

```python
df['Age_reciprocal'] = 1/df['Age']
plot_data(df, 'Age_reciprocal')
```

## Square Root Transformation

```python
df['Age_square'] = df.Age ** 1/2
plot_data(df, 'Age_square')
```

## Exponential Transdormation

```python
df['Age_expo'] = df.Age ** (1/1.2)
plot_data(df, 'Age_expo')
```

## BoxCOx Transformation
The Box-Cox transformation is defined as:

T(Y)=(Y exp()1)/

where Y is the response variable and  is the transformation parameter.  varies from -5 to 5. In the transformation, all values of  are considered and the optimal value for a given variable is selected.

```python
df['Age_Boxcox'],parameters=stat.boxcox(df['Age'])
plot_data(df,'Age_Boxcox')
```

```python
## Fare
df['Fare_log'] = np.log1p(df['Fare'])
plot_data(df, 'Fare_log')
```

```python
df['Fare_box_cox'], prams = stat.boxcox(df['Fare'] + 1)
plot_data(df, 'Fare_box_cox')
```
