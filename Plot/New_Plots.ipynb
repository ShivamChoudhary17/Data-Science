```python
import matplotlib.pyplot as plt

temp = [30, 32, 33, 28.5, 35, 29, 29]
ice_creams_count = [100, 115, 115, 75, 125, 79, 89]

plt.scatter(temp, ice_creams_count)
plt.title("Temperature vs. Sold ice creams")
plt.xlabel("Temperature")
plt.ylabel("Sold ice creams count")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('drugs_data.csv')
plt.plot(df['Age'], df['Sex'])
plt.title('Drugs Data')
plt.xlabel('Age')
plt.ylabel('Sex')
plt.show()
```

```python
df.hist(bins = 50)
```

```python
plt.figure(figsize=(16, 16))
plt.pie(df['Age'], autopct = "%.2f")
plt.axes().set_aspect("equal")
plt.show()
```
