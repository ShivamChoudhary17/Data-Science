```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df['smoker'].value_counts()
```

df.head()

### Corelation and Heatmap

- correlation- Helps to understand inter-relationship between any two feature

- Values are only found out when value are Integer or Floating

```python
df1 = df.corr()
```

## Heatmap


```python
sns.heatmap(df1)
```

## Joint plot

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

## Pair plot

- If we have more then two indepandent feature, also known as Scatterplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

## Dist plot

- Histogram

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'], kde = False, bins = 10) #Using kde y axis does't show in percentage
```

## Categorical  plot

### Plot categorical features NOT Numerics

- Boxplot
- Violin plot
- Countplot
- Bar plot

### COUNTPLOT always counts value for particuler Feature

```python
sns.countplot('day', data = df)
```

```python
sns.countplot(y = 'sex', data = df)
```

## Bar Plot
Always plot one with respect to other

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

## Box plot
According to percentile

```python
sns.boxplot('sex', 'total_bill', data = df)
```

```python
sns.boxplot('sex', 'total_bill', data = df, palette = 'Blues')
```

```python
sns.boxplot(data = df, orient = 'v') ## performs on all data set
```

## violin plot
Distribution of data with respect to kernal density estimation and the box plot

```python
sns.violinplot(x = 'total_bill', y = 'sex', data = df)
```
